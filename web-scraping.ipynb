# Web Scraping with python3.x

Supongamos que queremos leer el diario por internet. Lo primero que hacemos es abrir el navegador, escribir la URL del diario y apretar Enter para que aparezca la página del diario. Lo que ocurre en el momento en el que apretamos Enter es lo siguiente:

1. El navegador envía una solicitud a la URL pidiéndole información.
2. El servidor recibe la petición y procesa la respuesta.
3. El servidor envía la respuesta a la IP de la cual recibió la solicitud.
4. Nuestro navegador recibe la respuesta y la muestra formateada en pantalla.

Para hacer un scraper debemos hacer un programa que replique este flujo de forma automática para luego extraer la información deseada de la respuesta. Utilizaremos requests para realizar peticiones y recibir las respuestas y bs4 para parsear la respuesta y extraer la información.

Algunos links de utilidad 

* [Códigos de status HTTP](https://developer.mozilla.org/es/docs/Web/HTTP/Status)
* [Requests Docs](https://requests.kennethreitz.org/en/master/)

```python
from bs4 import BeautifulSoup
import lxml
```

```python
wp = p12.text
#soup = BeautifulSoup(wp, 'lxml') This line have something wrong!
soup = BeautifulSoup(wp, "html.parser")
```

```python
print(soup.prettify())
```

```python
find_ul = soup.find('ul', attrs={'class':'hot-sections'})
find_il = find_ul.find_all('li')
section = find_il[0]
```

```python
section.a.get_text()
```

```python
section.a.get('href')
```

```python
link_section = [section.a.get('href') for section in find_il]
```

```python
section_url = requests.get(link_section[0])
section_url.status_code
st = section_url.text 
soup_section = BeautifulSoup(st, "html.parser")
print(soup.prettify())
```
